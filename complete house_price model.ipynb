```python
import pandas as pd
```

```python
# Read the dataset

hp_df=pd.read_csv('house_price.csv')

hp_df
```

# ANOVA test

```python
from scipy.stats import f_oneway

sig_val=0.05
anova_corre_cols=[]
anova_noncorre_cols=[]

cat_cols=[col for col in hp_df.columns if hp_df[col].dtype=='object']

def anova(col):
    categor_list=list(hp_df[col].value_counts().index)
    
    result=f_oneway(*([hp_df[hp_df[col]==cat]['SalePrice'] for cat in categor_list]))
    
    return result[1],col

[anova(col)for col in cat_cols]
```

```python
p_values=[anova(col) for col in cat_cols]

for ele in p_values:
    if ele[0]<sig_val:
        anova_corre_cols.append(ele[1])
        
    if ele[0]>sig_val:
        anova_noncorre_cols.append(ele[1])    
        
print(anova_corre_cols)
 
print(anova_noncorre_cols)
        
```

# Drop the columns

```python
# we will drop y for train-test split

y=hp_df['SalePrice']
```

```python
# identifying columns with NAs greater than 50 percentage(there is no such rule)

na_df=pd.DataFrame({'col_name':hp_df.columns,'na_percent':hp_df.isnull().sum()/hp_df.shape[0]*100})

colu_grt50per_na=list(na_df[na_df['na_percent']>50]['col_name'])

colu_grt50per_na
```

```python
# identify non sense columns
# columns which have unique answer for full columns(location) or columns with each record as unique value(roll-no,ID)
# we will drop these non sense columns because we have no use with them

non_sense_col=[col for col in hp_df.columns if hp_df[col].nunique()==1 or hp_df[col].nunique()==hp_df[col].shape[0]]

non_sense_col
```

```python
# now drop all above columns(y,NA's>50,non_sense_columns)

cols_to_drop=colu_grt50per_na+non_sense_col+anova_noncorre_cols

cols_to_drop.append('SalePrice')
```

```python
# drop columns in data frame

hp_df.drop(columns=cols_to_drop,inplace=True)

hp_df
```

# Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(hp_df,y,train_size=0.8,random_state=42)
```

# Fill NA values

```python
# fill continuos NA's with mean
# fill categorical NA's with mode
```

```python
con_cols=[col for col in hp_df.columns if hp_df[col].dtype=='int64' or hp_df[col].dtype=='float64']
cat_cols=[col for col in hp_df.columns if hp_df[col].dtype=='object']
```

```python
# con_cols with mean

for col in con_cols:
    x_train[col].fillna(x_train[col].mean(),inplace=True)
    x_test[col].fillna(x_train[col].mean(),inplace=True)  # filling test as well as with train mean
    
# cat_cols with mode

for col in cat_cols:
    x_train[col].fillna(x_train[col].mode()[0],inplace=True)
    x_test[col].fillna(x_train[col].mode()[0],inplace=True)
    
```

# Scaling

```python
# scaling the continuos variables

from sklearn.preprocessing import StandardScaler
import numpy as np

std_scaler=StandardScaler()  # create a object

for col in con_cols:
    x_train[col]=std_scaler.fit_transform(np.array(x_train[col]).reshape(-1,1))
    x_test[col]=std_scaler.transform(np.array(x_test[col]).reshape(-1,1))
    
```

```python
# encoding for categorical variables

oe_train=pd.get_dummies(x_train[cat_cols])
oe_test=pd.get_dummies(x_test[cat_cols])
```

```python
print(oe_train.shape)

print(oe_test.shape)
```

```python
# after encoding the columns are diff (238,202) for train and test
# so we have to make them equal
# first merge them using allign(see in web -->pandas join two data frames)

oe_train_final,oe_test_final=oe_train.align(oe_test,join='inner',axis=1)
```

```python
print(oe_train_final.shape)

print(oe_test_final.shape)  # now both columns are equal
```

```python
# as we have divided the con_cols and cat_cols to fill NA vales and to scale them
# now to train the model we have to merge them using pandas 'concat'

x_train_final=pd.concat([x_train[con_cols],oe_train_final],axis=1)
x_test_final=pd.concat([x_test[con_cols],oe_test_final],axis=1)
```

```python
x_train_final.shape[1]
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()

lin_reg.fit(x_train_final,y_train)  # the model had been trained

train_pred=lin_reg.predict(x_train_final)

test_pred=lin_reg.predict(x_test_final)  # the test data has been predicted
```
